```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import value_counts
# в df представлены виды покемонов и их супер силы
path = 'Pokemon.csv'
df = pd.read_csv(path, delimiter=',',index_col= '#')
df.head()
```

```python
top = pd.DataFrame(df['Type 1'].value_counts())
top_top = pd.DataFrame(df['Type 1'].value_counts().head(3)) #получаем топ 3 часто используемых покемонов
top_top
```

```python
top_small = pd.DataFrame(df['Type 1'].value_counts().nsmallest(3)) #получаем топ 3 редко используемых
top_small
#теперь построим общий график покемонов первого типа
```

```python
sns.barplot(data = top, x ='Type 1', y='count', hue = 'Type 1')
plt.xticks(rotation=45)
plt.xlabel('Type of pokemon')
plt.ylabel('Count of pokemon')
```

```python
top3 = df[(df['Type 1'] == 'Water') | (df['Type 1'] == 'Normal') | (df['Type 1'] =='Grass')]
# делаю выборку по трем часто используемых покемонов Типа 1
top3
```

```python
top3_leg = top3[top3['Legendary'] == False]
# и выбираю тех, кто не относится к легендарным
top3_leg
```

```python
median_strength = top3_leg['Total'].median()
median_strength
```
